```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
matches = pd.read_csv('WorldCupMatches.csv')
```

```python
df_cleaned = matches.dropna(subset=['Year']).copy()
df_cleaned = df_cleaned.drop_duplicates(subset=['MatchID'])
```

```python
df_cleaned['Year'] = df_cleaned['Year'].astype(int)
df_cleaned['Home Team Goals'] = df_cleaned['Home Team Goals'].astype(int)
df_cleaned['Away Team Goals'] = df_cleaned['Away Team Goals'].astype(int)
df_cleaned['Half-time Home Goals'] = df_cleaned['Half-time Home Goals'].fillna(0).astype(int)
df_cleaned['Half-time Away Goals'] = df_cleaned['Half-time Away Goals'].fillna(0).astype(int)
df_cleaned['Attendance'] = df_cleaned['Attendance'].fillna(df_cleaned['Attendance'].median())
```

```python
df_cleaned['Home_Win'] = (df_cleaned['Home Team Goals'] > df_cleaned['Away Team Goals']).astype(int)
```

```python
df_cleaned['Is_Knockout'] = df_cleaned['Stage'].apply(lambda x: 0 if 'Group' in str(x) else 1)
```

```python
print(f"Data cleaned successfully! Number of unique matches remaining for training: {len(df_cleaned)}")
```

```python
sns.set_theme(style='whitegrid')
corr_cols = ['Year', 'Attendance', 'Half-time Home Goals', 'Half-time Away Goals', 'Is_Knockout', 'Home_Win']
```

```python
plt.figure(figsize=(7, 5))
sns.heatmap(df_cleaned[corr_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of World Cup Match Features')
plt.tight_layout()
plt.savefig('heatmap.png')
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.boxplot(x='Home_Win', y='Half-time Home Goals', data=df_cleaned)
plt.title('Boxplot of Half-time Home Goals by Match Outcome')
plt.xlabel('Home Team Win (0: No, 1: Yes)')
plt.ylabel('Half-time Home Goals')
plt.tight_layout()
plt.savefig('boxplot.png')
plt.show()
```

```python
p_coeff, p_value = pearsonr(df_cleaned['Half-time Home Goals'], df_cleaned['Home_Win'])
print(f"\n Pearson Correlation (Half-time Goals & Home_Win): {p_coeff:.4f}")
print(f" p-value: {p_value:.4e}\n")
```

```python
le_home = LabelEncoder()
le_away = LabelEncoder()
df_cleaned['Home_Team_Enc'] = le_home.fit_transform(df_cleaned['Home Team Name'])
df_cleaned['Away_Team_Enc'] = le_away.fit_transform(df_cleaned['Away Team Name'])
```

```python
features = ['Year', 'Attendance', 'Half-time Home Goals', 'Half-time Away Goals', 'Is_Knockout', 'Home_Team_Enc', 'Away_Team_Enc']
X = df_cleaned[features]
y = df_cleaned['Home_Win']
```

```python
scaler = StandardScaler()
X_scaled = X.copy()
X_scaled[['Year', 'Attendance', 'Half-time Home Goals', 'Half-time Away Goals']] = scaler.fit_transform(X[['Year', 'Attendance', 'Half-time Home Goals', 'Half-time Away Goals']])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
print(f"Data split completed successfully!")
print(f"Training set size: {X_train.shape[0]} matches")
print(f"Testing set size: {X_test.shape[0]} matches\n")
```

```python
lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(X_train, y_train)
lr_preds = lr_model.predict(X_test)
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [2, 5, 10]
}
```

```python
grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
best_rf_model = grid_search.best_estimator_
rf_preds = best_rf_model.predict(X_test)
print(f"Best Random Forest Parameters discovered: {grid_search.best_params_}\n")
```

```python
def calculate_metrics(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    return [acc, prec, rec, f1]

lr_metrics = calculate_metrics(y_test, lr_preds)
rf_metrics = calculate_metrics(y_test, rf_preds)
```

```python
comparison_df = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'Logistic Regression': lr_metrics,
    'Tuned Random Forest': rf_metrics
})
print("--- MODEL PERFORMANCE COMPARISON TABLE ---")
print(comparison_df.to_string(index=False))
```

```python
plt.figure(figsize=(8, 5))
comparison_melted = pd.melt(comparison_df, id_vars=['Metric'], var_name='Model', value_name='Value')
sns.barplot(x='Metric', y='Value', hue='Model', data=comparison_melted, palette='muted')
plt.title('Performance Metrics Comparison')
plt.ylim(0, 1.0)
plt.ylabel('Score')
plt.tight_layout()
plt.savefig('performance_comparison.png')
plt.show()
```

```python
style = {'description_width': 'initial'}
```

```python
year_widget = widgets.IntSlider(value=2026, min=1930, max=2026, step=4, description='Tournament Match Year:', style=style)
attendance_widget = widgets.IntSlider(value=45000, min=2000, max=100000, step=1000, description='Stadium Attendance:', style=style)
knockout_widget = widgets.Dropdown(options=[('Group Stage', 0), ('Knockout Round', 1)], value=0, description='Tournament Stage:', style=style)

home_teams = sorted(df_cleaned['Home Team Name'].unique())
away_teams = sorted(df_cleaned['Away Team Name'].unique())

home_team_widget = widgets.Dropdown(options=home_teams, value='Brazil', description='Designated Home Team:', style=style)
away_team_widget = widgets.Dropdown(options=away_teams, value='Italy', description='Designated Away Team:', style=style)

ht_home_widget = widgets.IntSlider(value=1, min=0, max=5, description='Half-Time Home Team Goals:', style=style)
ht_away_widget = widgets.IntSlider(value=0, min=0, max=5, description='Half-Time Away Team Goals:', style=style)

predict_btn = widgets.Button(description="Predict Match Outcome", button_style='success')
output_area = widgets.Output()
```

```python
def on_predict_click(b):
    with output_area:
        output_area.clear_output()

        h_team = home_team_widget.value
        a_team = away_team_widget.value

        try:
            h_enc = le_home.transform([h_team])[0]
            a_enc = le_away.transform([a_team])[0]
        except ValueError:
            h_enc = 0
            a_enc = 0

        raw_data = pd.DataFrame([{
            'Year': year_widget.value,
            'Attendance': attendance_widget.value,
            'Half-time Home Goals': ht_home_widget.value,
            'Half-time Away Goals': ht_away_widget.value
        }])
```

```python
def on_predict_click(b):
    with output_area:
        output_area.clear_output()

        h_team = home_team_widget.value
        a_team = away_team_widget.value

        try:
            h_enc = le_home.transform([h_team])[0]
            a_enc = le_away.transform([a_team])[0]
        except ValueError:
            h_enc = 0
            a_enc = 0

        raw_data = pd.DataFrame([{
            'Year': year_widget.value,
            'Attendance': attendance_widget.value,
            'Half-time Home Goals': ht_home_widget.value,
            'Half-time Away Goals': ht_away_widget.value
        }])

        scaled_values = scaler.transform(raw_data[['Year', 'Attendance', 'Half-time Home Goals', 'Half-time Away Goals']])

        final_input_df = pd.DataFrame([{
            'Year': scaled_values[0][0],
            'Attendance': scaled_values[0][1],
            'Half-time Home Goals': scaled_values[0][2],
            'Half-time Away Goals': scaled_values[0][3],
            'Is_Knockout': knockout_widget.value,
            'Home_Team_Enc': h_enc,
            'Away_Team_Enc': a_enc
        }])

        pred = lr_model.predict(final_input_df)[0]
        prob = lr_model.predict_proba(final_input_df)[0][1]

        if knockout_widget.value == 1:
            stage_text = "Knockout Round (Elimination)"
        else:
            stage_text = "Group Stage (Round-Robin)"

        print("\n" + "╔" + "═"*58 + "╗")
        print(f"║ {'MATCH CONTEXT & ENVIRONMENT DETAILS':^56} ║")
        print("╠" + "═"*58 + "╣")
        print(f"║ Fixture: {h_team} vs {a_team:<36} ║")
        print(f"║ Tournament Year: {year_widget.value:<39} ║")
        print(f"║ Stadium Attendance: {attendance_widget.value:<34,} ║")
        print(f"║ Match Stage: {stage_text:<39} ║")
        print("╠" + "═"*58 + "╣")
        print(f"║ {'MID-GAME STATUS (HALF-TIME SCORE)':^56} ║")
        print("╠" + "═"*58 + "╣")
        print(f"║ Half-Time Score: {h_team} {ht_home_widget.value} - {ht_away_widget.value} {a_team:<22} ║")
        print("╠" + "═"*58 + "╣")
        print(f"║ {'PREDICTIVE ANALYTICS ENGINE OUTPUT':^56} ║")
        print("╠" + "═"*58 + "╣")

        if pred == 1:
            outcome_text = f"SUCCESS: {h_team.upper()} VICTORY PREDICTED!"
            print(f"║ {outcome_text:^56} ║")
            print(f"║ Definition: The Home Team is highly likely to win.      ║")
        else:
            outcome_text = "DRAW OR AWAY TEAM WIN PREDICTED!"
            print(f"║ {outcome_text:^56} ║")
            print(f"║ Definition: The Home Team is likely to lose or draw.    ║")

        print(f"║ Estimated Win Probability for {h_team}: {prob*100:.2f}%{' ':<11} ║")
        print("╚" + "═"*58 + "╝\n")
```

```python
predict_btn.on_click(on_predict_click)
display(year_widget, attendance_widget, knockout_widget, home_team_widget, away_team_widget, ht_home_widget, ht_away_widget, predict_btn, output_area)
```
